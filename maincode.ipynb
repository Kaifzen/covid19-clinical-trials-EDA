# Data Cleaning


In this step, we analyze and clean the dataset by identifying columns with missing values and applying appropriate strategies to handle them.

```python
# Percentage of missing values
missing_percent = df.isnull().mean() * 100
missing_percent.sort_values(ascending=False)
```

'Results First Posted' and 'Study Documents' were missing in over 95% of the entries, making them unfit for imputation and ultimately dropped.



```python
#Dropping columns with large number of missing values
df.drop(columns=['Results First Posted', 'Study Documents', 'URL', 'NCT Number'], inplace=True)
```

```python
#Handling categorical missing data
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(f"Missing {col}", inplace=True)
```

```python
#Handling numerical missing data
df['Enrollment'] = df['Enrollment'].fillna(df['Enrollment'].median())
```

```python
#Identifiy and convert relevant columns
date_cols = ['Start Date', 'Primary Completion Date', 'Completion Date', 'First Posted', 'Last Update Posted']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

'Acronym' and 'Phases' had moderate missingness but are still useful for categorical analysis, so we filled them with "Missing <column>" to preserve information while avoiding NaNs.

For other categorical columns like 'Interventions', 'Gender', and 'Locations', we also filled missing entries with placeholder text.

The numeric column 'Enrollment' showed a strong right skew, so we imputed missing values using the median, a robust measure less sensitive to outliers.

```python
#Extract country from Locations
# If Locations exists and not already split
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1].strip() if pd.notnull(x) else 'Missing Country')
```

```python
#Create trial duration feature
df['Trial Duration (Days)'] = (df['Completion Date'] - df['Start Date']).dt.days
```

We also engineered new, meaningful features:

Country — extracted from the 'Locations' column (by splitting the last part of the string)

Trial Duration (Days) — calculated by subtracting 'Start Date' from 'Completion Date' for each study

```python
#Final check
df.info()
df.isnull().sum()  # Final sanity check
```

```python
#Saving the cleaned dataset
df.to_csv('cleaned_covid_clinical_trials.csv', index=False)
```

Dataset is now fully clean and ready for exploratory analysis

Bivariate Analysis

This bivariate analysis explores how clinical trial phases are distributed across different statuses (e.g., Recruiting, Completed). Instead of using a stacked bar chart, we opted for a grouped bar plot with labeled counts for more clarity and comparison.

```python
# Crosstab of Status and Phases
# Step 1: Simplify phases
def simplify_phase(p):
    p = str(p)
    if 'Phase 1' in p and 'Phase 2' in p:
        return 'Phase 1/2'
    elif 'Phase 2' in p and 'Phase 3' in p:
        return 'Phase 2/3'
    elif 'Phase 1' in p:
        return 'Phase 1'
    elif 'Phase 2' in p:
        return 'Phase 2'
    elif 'Phase 3' in p:
        return 'Phase 3'
    elif 'Phase 4' in p:
        return 'Phase 4'
    elif 'Early' in p:
        return 'Early Phase 1'
    elif 'Not Applicable' in p:
        return 'Not Applicable'
    elif 'Missing' in p:
        return 'Missing Phase'
    else:
        return 'Other'

df['Simplified Phase'] = df['Phases'].apply(simplify_phase)

```

```python
# Step 2: Crosstab and order

# Count phase distribution by status
simplified_counts = df.groupby(['Status', 'Simplified Phase']).size().reset_index(name='Count')

# Sort statuses by overall frequency
status_order = df['Status'].value_counts().index.tolist()
phase_order = ['Missing Phase', 'Not Applicable', 'Phase 1', 'Phase 1/2', 'Phase 2', 'Phase 2/3', 'Phase 3', 'Phase 4', 'Early Phase 1']

# Plot grouped bars with labels
plt.figure(figsize=(16,6))
ax = sns.barplot(data=simplified_counts,
                 x='Status', y='Count', hue='Simplified Phase',
                 order=status_order, hue_order=phase_order)

# Add value labels on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=8)

plt.title('Status vs. Simplified Trial Phases')
plt.xlabel('Trial Status')
plt.ylabel('Number of Trials')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Phase', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

🟩 The "Recruiting" category has the highest number of trials, dominated by Missing Phase, Not Applicable, and early phases.

🟨 Completed trials show strong presence in Phase 2, Phase 3, and Phase 4, indicating successful progression of mid-to-late-stage studies.

🚫 Trial statuses such as Withdrawn, Terminated, and Suspended have noticeably fewer entries — often linked with Phase 2 or earlier.

📌 Labels above each bar segment provide immediate visibility into trial distribution across phases and statuses — enhancing interpretability.
